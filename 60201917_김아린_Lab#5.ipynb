```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
import matplotlib as mpl



iris = load_iris()


iris.target
iris
```

```python
iris.data
```

```python

frame = pd.DataFrame(iris.data, columns=['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width(cm)'])

frame['target'] = iris.target
frame['targetName'] = np.where(frame["target"].values == 0, "setosa", "versicolor")
frame.loc[frame['target'] == 2 , 'targetName'] = "virginica"

frame.head()
```

```python
plt.figure(figsize = (10,10)) # 
plt.subplot(2,1,1)
x = frame['sepal length (cm)'].iloc[:49]
y = frame['sepal width (cm)'].iloc[:49]
plt.scatter(x, y, marker = 'o', color = 'red', label = 'setosa')
x2 = frame['sepal length (cm)'].iloc[50:99]
y2 = frame['sepal width (cm)'].iloc[50:99]
plt.scatter(x2, y2, marker = 'x', color = 'blue', label = 'versicolor')
x3 = frame['sepal length (cm)'].iloc[100:]
y3 = frame['sepal width (cm)'].iloc[100:]
plt.scatter(x3, y3, marker = '^', color = 'green', label = 'virginica')
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.legend()

plt.subplot(2,1,2)
x = frame['petal length (cm)'].iloc[:49]
y = frame['petal width(cm)'].iloc[:49]
plt.scatter(x, y, marker = 'o', color = 'red', label = 'setosa')
x2 = frame['petal length (cm)'].iloc[50:99]
y2 = frame['petal width(cm)'].iloc[50:99]
plt.scatter(x2, y2, marker = 'x', color = 'blue', label = 'versicolor')
x3 = frame['petal length (cm)'].iloc[100:]
y3 = frame['petal width(cm)'].iloc[100:]
plt.scatter(x3, y3, marker = '^', color = 'green', label = 'virginica')
plt.xlabel('petal length')
plt.ylabel('petal width')
plt.legend()



plt.show()
```
