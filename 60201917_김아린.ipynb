```python
import numpy as np


def main() :
    Score = np.random.randint(100, size=(10, 4))
    PrintScore(Score)
    MathScore(Score)
    TotalScore(Score)
    BooleanScore(Score)
    EngScore(Score)
    

def PrintScore(Score) :
    print(Score) 
    
    
def MathScore(Score) :
    for i in range(10) :
        MathScore = Score[i, 2:3]
        print(i, '번째 학생의 수학 점수 :', MathScore)
        
def TotalScore(Score) :
    for i in range(10) :
        TotalScore = 0
        for j in range(4) :
            TotalScore += Score[i][j]
        print(i, '번째 학생의 과목 총합 점수 :', TotalScore)
        
        
def BooleanScore(Score) :
    for i in range(10) :
        if i % 2 != 0 :
            data = Score[i:i+1, :]
            print(i, '번째 학생의 과학, 수학, 국어, 영어 점수 :', np.flip(data, axis=1))
            
    
        
        
def MaxScore(Score) :
    subject = ['국어', '영어', '수학', '과학']
    for i in range(4) :
        maxScore = max(Score[:, i:i+1])
        print(subject[i], '과목의 최고 점수 :', maxScore)
        
        
        
        
        
def EngScore(Score) :
    for i in range(10) :
        data = Score[i][1] < 90
        if data == True :
            print(i, '번째 학생의 영어 점수는 90점 미만이며,', Score[i][1], '점 입니다.')

        
    
    
    
main()
    
```
