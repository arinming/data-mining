```python
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
import numpy as np
from pandas import Series, DataFrame
import pandas as pd
import seaborn as sns

data = pd.read_csv("nba.csv")
data
```

```python
data_team = data[(data['Team']=="Houston Rockets")|(data['Team']=="Los Angeles Lakers")]
data_team[['Age', 'Salary', 'Team', 'Position']]
```

```python
sns.set_theme(style="whitegrid")
g1 = sns.boxplot(x="Age", y="Salary", hue="Position", data=data_team)
```

# 분석 결과 해석

30세가 연봉이 가장 높다.
C 포지션이 연봉 범위가 가장 크며, 연봉을 많이 받는다.

    

```python
sns.jointplot(x="Age", y="Salary", data=data_team, kind='reg');
```

# 분석 결과 해석

나이와 월급의 상관관계가 나타나있다.
20-25세 정도 까지는 상관관계가 어느정도 일치하게 데이터가 분포되어있지만,
그 이상의 나이부터는 상관관계에 멀게 데이터가 분포되어있는 것을 확인할 수 있다.

    

```python
with sns.axes_style('white'):
    g = sns.catplot(x="Age", data=data_team, aspect=4.0, kind='count',
                       hue='Position')
    g.set_ylabels('count')
```

# 분석 결과 해석

PF 포지션은 주로 21-25세 정도에 많이 분포되어 있다. C 포지션은 27-30세 정도에 많이 분포되어 있다. SF 포지션과 PG 포지션, SG 포지션은 전체적으로 골고루 분포되어 있다.
