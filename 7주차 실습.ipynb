```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
import matplotlib as mpl

dia = pd.read_csv("diabetes.csv")
dia_age = pd.Series(dia["Age"])

dia_age_sort = pd.Series(sorted(dia_age))

plt.figure()
plt.bar(dia_age_sort.index,  dia_age_sort.values)
plt.show()
```

```python
def dia_max(x) :
    return pd.Series([dia["Age"].min(), x.max()], index=['min', 'max'])

dia_max(dia)
```
