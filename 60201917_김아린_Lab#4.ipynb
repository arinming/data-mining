```python
%matplotlib inline
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt




def main() :
    Score = np.random.randint(100, size=(10, 4))
    plt.style.use('classic')
    PrintScore(Score)
    Q1(Score)
    Q2(Score)
    Q3(Score)
    Q4(Score)

    
    
def PrintScore(Score) :
    print(Score) 
    
    
# 1 
    
def Q1(Score):
    plt.xlabel("English")
    plt.ylabel("Korea");
    x = Score[:,1]
    y = Score[:,0]
    plt.scatter(x, y, marker='o')
    plt.show()
    
# 2 
    
def Q2(Score):
    fig = plt.figure()
    fig.subplots_adjust(hspace=0.4, wspace=0.4)
    sub = ["Korea", "English", "Math", "Science"]
    for i in range(1, 5):
        ax = fig.add_subplot(2, 2, i)
        data = Score[:,i-1]
        plt.ylabel("Count")
        plt.xlabel(sub[i-1])
        plt.hist(data)
    plt.show()

        
# 3 

def Q3(Score):
    a = 0
    b = 0
    c = 0
    d = 0
    for i in range(10) :
        if Score[i][0] >= 90 :
            a += 1
        elif Score[i][0] >= 80 and Score[i][0] < 90 :
            b += 1
        elif Score[i][0] >= 70 and Score[i][0] < 80 :
            c += 1
        else :
            d += 1
    arrlist = [int(a),int(b),int(c),int(d)]
    label=['A','B','C', 'D']
    plt.pie(arrlist,labels=label)
    plt.show()
    
# 4 

def Q4(Score):
    plt.boxplot(Score)
    plt.xticks([1,2,3,4],["Korea", "English", "Math", "Science"])
    plt.show()


main()
```
