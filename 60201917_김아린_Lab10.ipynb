```python
import seaborn as sns
import altair as alt

mpg=sns.load_dataset('mpg')
```

```python



alt.Chart(mpg).mark_bar().encode(
    x='cylinders:O',
    y='sum(horsepower):Q',
    color='cylinders:N',
    column='origin:N'
)
```

```python
pl=sns.load_dataset('flights')
```

```python
pl
```

```python
pl_1 = pl.loc[pl["month"] == "Jan", ["year", "month", "passengers"]]
pl_1
```

```python
pl_2 = pl.loc[pl["month"] == "Aug", ["year", "month", "passengers"]]
pl_2
```

```python

alt.Chart(pl_1).mark_line(interpolate='step-after').encode(
    x='year',
    y='passengers'
) | alt.Chart(pl_2).mark_line(interpolate='step-after').encode(
    x='year',
    y='passengers'
)
```
